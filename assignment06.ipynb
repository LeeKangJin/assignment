```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv

num     = 201
std     = 20
a       = 2 
b       = 10

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-100,100,num) # noisy x 
y1      = a * x + nn * std + b # noisy y domain
```

# Noisy X,Y1 Grape

```python
plt.plot(x,y1)   # plot noisy_x & y
plt.show()
```

# Least Square Method

# Sudo Inverse A 
                    

```python
# Sudo Inverse A = (A_Transpose * A) Inverse * A_Transpose *b


one_matrix = np.ones(num) # init

A = np.column_stack((x,one_matrix)) # [ [x1,1], [x2,1] , [x3,1] ....]
A_transpose = A.transpose()

Sudo_InverseA = np.matmul(A_transpose,A)
Sudo_InverseA = np.linalg.inv(Sudo_InverseA)
Sudo_InverseA = np.matmul(Sudo_InverseA,A_transpose)
```

# b (Initialize)


# Clean data = A(SudoInverse) * b

```python
data_size = np.shape(y1)[0]
b = y1.reshape((data_size,1))
soluation = np.matmul(Sudo_InverseA,b)

```

```python
y2 = np.matmul(A,soluation)
plt.plot(x,y2)
plt.show()
```

# Plot Clean Line & Noisy

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```
