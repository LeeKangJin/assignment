# import packages for plotting graphs and manipulating data:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # Axes3D is 3D package 
```

# define my function: $f(x,y) = sin(x)*y+cos(y)*x$

```python
def myFunctions(x, y):
    f = np.sin( x ) * y + np.cos( y ) * x
    return f
```

# define the derivative Function of x and y:
\begin{align}
\frac{\partial f}{\partial x}(x,y) &= y*cos(x)+cos(y),  \\
\frac{\partial f}{\partial y}(x,y) &= sin(x) -x*sin(y)
\end{align}

```python
def derivativeXFunction(x, y):
    dfx = y * np.cos( x ) + np.cos( y )
    return dfx
```

```python
def derivativeYFunction(x, y):
    dfy = np.sin( x ) - x * np.sin( y )
    return dfy
```

# define the Taylor approximation function:
\begin{align}
      ta(x) = \frac{\partial f}{\partial x}(a,b)(x-f(a,b))  \\
      ta(y) = \frac{\partial f}{\partial y}(a,b)(y-f(a,b))  \\
      t(x,y) = ta(x) + ta(y) + f(a,b)  \\
\end{align}

```python
def taylorApproximation(x, y, a, b):

    taX = np.dot(derivativeXFunction( a, b ),x - myFunctions( a, b ))
    taY = np.dot( derivativeYFunction( a, b ), y - myFunctions( a, b ))
    ta = taX +taY +myFunctions(a,b)

    return ta
```

# define the domain of the function :
$f\colon X\to Y$: $ \{ (x,y) \mid x\in[-10:1:10], y\in[-10:1:10] \}$

```python
x = np.arange( -10, 1, 10 )
y = np.arange( -10, 1, 10 )
X, Y = np.meshgrid( x, y )
```

# Pick 3 points in the domain    :
\begin{align}
P1 &= (0,0),  \\
P2 &= (-10,1), \\
P3 &= (-5,5)
\end{align}

# Plot the graph of defined function

```python
case = '1'                                         # 1 = original function
if case is '1':
    fig = plt.figure()                             # 3D environment setting
    axes = Axes3D( fig )                           # 3D environment setting
    axes.set_title( "Original function" )          # title name
    axes.plot_surface( X, Y, myFunctions( X, Y ) )
    plt.show()
```

# Plot the graph of Taylor approximation  for the given points : [0,0]


```python
case ='2'                                          # case of given point [0,0]
if case is '2':
    fig = plt.figure()                             # 3D environment setting
    axes = Axes3D( fig )                           # 3D environment setting
    axes.set_title( "Talyor Approximation" )
    axes.plot_surface( X, Y, taylorApproximation(X,Y,0,0))
    plt.show()
```
